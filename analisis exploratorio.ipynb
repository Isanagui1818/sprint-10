```python
import os
import pandas as pd

# Obtener todos los años únicos de los archivos en la carpeta
años_disponibles = set(archivo[:4] for archivo in os.listdir(carpeta_origen) if archivo.endswith(".csv") and archivo[:4].isdigit())

# Iterar sobre cada año disponible
for año_filtrar in años_disponibles:
    # Crear una lista para almacenar los dataframes de este año
    dataframes = []
    
    # Iterar sobre los archivos en la carpeta origen
    for archivo in os.listdir(carpeta_origen):
        # Verificar si el archivo empieza con el año actual y es un CSV
        if archivo.startswith(año_filtrar) and archivo.endswith(".csv"):
            # Leer el archivo CSV y añadirlo a la lista de dataframes
            ruta_archivo = os.path.join(carpeta_origen, archivo)
            df = pd.read_csv(ruta_archivo, low_memory=False)
            dataframes.append(df)
    
    # Combinar todos los dataframes del año actual en uno solo
    if dataframes:  # Verificar que la lista no esté vacía
        dataframe_final = pd.concat(dataframes, ignore_index=True)
        
        # Crear dinámicamente una variable global con el nombre del año
        nombre_variable = f"dataframe_final_{año_filtrar}"
        globals()[nombre_variable] = dataframe_final
        
        print(f"Archivos del año {año_filtrar} combinados exitosamente en un dataframe llamado '{nombre_variable}'.")
    else:
        print(f"No se encontraron archivos para el año {año_filtrar}.")
```

```python
import plotly.express as px
import geopandas as gpd
import pandas as pd
import fiona

# Cargar el archivo GeoDataFrame con información de los distritos
try:
    districts = gpd.read_file(distritos_geojson)
except fiona.errors.DriverError:
    print(f"Error: No se pudo abrir el archivo {distritos_geojson}. Asegúrate de que sea un GeoJSON válido.")
    exit()

# *** OPCIONES PARA MEJORAR LA PRECISIÓN DE LA GEOMETRÍA ***
# Reducir el umbral de simplificación
districts['geometry'] = districts['geometry'].apply(lambda x: x.simplify(0.0001))

# Asegúrate de usar el sistema de coordenadas WGS84 (EPSG:4326)
districts = districts.to_crs('EPSG:4326')

# Definir una lista de colores HEX
colores_hex = [
    '#000000',  # Negro Ciutat Vella
    '#377EB8',  # Azul Eixample
    '#DB6D00',  # Marrón Gracia
    '#490092',  # Morado
    '#F100F1',  # Naranja Les corts
    '#FFFF33',  # Amarillo Nou Barris
    '#A65628',  # Marrón
    '#F781BF',  # Rosa Sant-Martí
    '#FFB900',  # Dorado Sants-Montjuïc
    '#66A61E'   # Verde oscuro
    # Añade más colores hex si tienes más distritos
]

def hex_to_rgba(hex_color, alpha):
    """Convierte un color HEX a formato RGBA con la transparencia especificada."""
    hex_color = hex_color.lstrip('#')
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)
    return f'rgba({r}, {g}, {b}, {alpha})'

# Convertir los colores HEX a RGBA con transparencia 0.3
colores_translucidos_hex = [hex_to_rgba(hex_color, 0.3) for hex_color in colores_hex[:len(districts)]]

# Crear el mapa base de los distritos
fig = px.choropleth_map(
    districts,
    geojson=districts.geometry.__geo_interface__,
    locations=districts.index,
    color='NOM',
    category_orders={"NOM": districts['NOM'].sort_values().unique().tolist()},
    color_discrete_sequence=colores_translucidos_hex,
    center={"lat": 41.3851, "lon": 2.1734},
    zoom=11,
    title="Estaciones Bicing geolocalizadas en Barcelona",
    labels={'NOM': 'Nombre del Distrito'}
)

# --- Añadir las estaciones de Bicing ---
# ASUMIENDO QUE 'Info_bicing_csv' ES TU DATAFRAME CARGADO CON TODOS LOS DATOS
# Y TIENE LAS COLUMNAS 'lat', 'lon' Y 'name'

fig.add_trace(px.scatter_map(Info_bicing_csv, lat="lat", lon="lon",
                             zoom=12, size_max=9,
                             color_discrete_sequence=["#E9514C"],
                             opacity=0.9,
                             hover_data={'lat': False, 'lon': False, 'station_id': True}  # Solo especificamos 'name'
                             ).data[0])

# Modificar el tamaño del layout
fig.update_layout(
    margin={"r":0,"t":30,"l":0,"b":0},
    width=1000,  # Ajusta el ancho a tu preferencia
    height=700   # Ajusta la altura a tu preferencia para hacerlo más cuadrado
)

fig.show()
```

```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import fiona

# Cargar el archivo GeoDataFrame con información de los distritos
try:
    districts = gpd.read_file(distritos_geojson)
except fiona.errors.DriverError:
    print(f"Error: No se pudo abrir el archivo {distritos_geojson}. Asegúrate de que sea un GeoJSON válido.")
    exit()

# Asegúrate de usar el sistema de coordenadas WGS84 (EPSG:4326)
districts = districts.to_crs('EPSG:4326')

# *** ASUMIENDO QUE dataframe_consolidado YA ESTÁ CARGADO Y TIENE LAS COLUMNAS: ***
# 'station_id', 'lat', 'lon', 'last_reported' (timestamp), 'num_docks_available'

# Convertir la columna 'last_reported' a datetime
dataframe_consolidado['timestamp'] = pd.to_datetime(dataframe_consolidado['last_reported'], unit='s')

# Extraer el año y el mes
dataframe_consolidado['year'] = dataframe_consolidado['timestamp'].dt.year
dataframe_consolidado['month'] = dataframe_consolidado['timestamp'].dt.month

# Obtener la ubicación única de cada estación y su distrito
estaciones_unicas = dataframe_consolidado[['station_id', 'lat', 'lon']].drop_duplicates(subset=['station_id']).copy()
estaciones_unicas = gpd.GeoDataFrame(estaciones_unicas, geometry=gpd.points_from_xy(estaciones_unicas.lon, estaciones_unicas.lat), crs="EPSG:4326")
estaciones_con_distrito = gpd.sjoin(estaciones_unicas, districts[['NOM', 'geometry']], how="inner", predicate='within')
dataframe_consolidado_merged = pd.merge(dataframe_consolidado, estaciones_con_distrito[['station_id', 'NOM']], on='station_id', how='left')
dataframe_consolidado_merged.rename(columns={'NOM': 'Distrito'}, inplace=True)

# Calcular la media de 'num_docks_available' por mes, distrito y año
monthly_avg_docks_by_district_year = dataframe_consolidado_merged.groupby(['Distrito', 'year', 'month'])['num_docks_available'].mean().reset_index()

# Crear el gráfico de líneas con diferenciación por distrito y año
fig = px.line(monthly_avg_docks_by_district_year,
              x='month',
              y='num_docks_available',
              color='Distrito',
              line_dash='year',  # Diferenciar las líneas por año con estilos de línea
              title='Media Mensual de Docks Disponibles por Distrito (2020-2024)',  # Ajusta los años según tus datos
              labels={'month': 'Mes', 'num_docks_available': 'Media de Docks Disponibles', 'Distrito': 'Distrito', 'year': 'Año'},
              markers=True)

# Personalizar las marcas del eje x para mostrar nombres de los meses
fig.update_xaxes(tickvals=list(range(1, 13)),
                 ticktext=['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
                           'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

fig.show()
```
